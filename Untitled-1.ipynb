```python
#importing lib
import pandas as pd
import numpy as np
import seaborn as sns

import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import numpy as np
import matplotlib
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)

pd.options.mode.chained_assignment = None

#reading data
df=pd.read_csv(r'D:\DATA ANALYST PORTFOLIO\movies.csv')
```

```python
df.head()
```

```python
# missing data
for col in df.columns:
    missing_data= np.mean(df[col].isnull())
    print('{}-{}%'.format(col,missing_data))            
```

```python
#dataypes of columns
df.dtypes
```

```python
#changing dataypes
df['budget']=df['budget'].fillna(0).astype('int64')
df['gross']=df['gross'].fillna(0).astype('int64')
df.dropna()
```

```python
df['month_correct']=df['released'].astype('str').str[:4]
df
```

```python
df.sort_values(by=['gross'], inplace=False,ascending=False)
df.head()
```

```python
#remove duplicates
df['company'].drop_duplicates().sort_values(ascending=False)
```

```python
#scatter plots with budget vs gross

plt.scatter(x=df['budget'],y=df['gross'])
plt.title('Budget vs Gross earnings')
plt.xlabel('Gross earnings')
plt.ylabel('Budget for film')
plt.show()
```

```python
#plot budget vs gross using seaborn to show correlation.
sns.regplot(x='budget',y='gross',data=df,scatter_kws={"color":"black"},line_kws={"color":"blue"})
```

```python
#correlation
df[['budget', 'gross', 'score', 'votes', 'runtime','year']].corr()
```

```python
#3 methods to find corr (pearson,kendall,spearman)
df[['budget', 'gross', 'score', 'votes', 'runtime','year']].corr(method='pearson')
```

```python
df[['budget', 'gross', 'score', 'votes', 'runtime','year']].corr(method='kendall')
```

```python
df[['budget', 'gross', 'score', 'votes', 'runtime','year']].corr(method='spearman')
```

```python
#high corr between gross and budget
correlation_matrix=df[['budget', 'gross', 'score', 'votes', 'runtime','year']].corr(method='pearson')
sns.heatmap(correlation_matrix, annot=True)
plt.title('Correlation matrice for  numeric features')
plt.xlabel('Movie features')
plt.ylabel('Movie features')
plt.show()
```

```python
#since data in company is not numeric. so we'll convert it into numeric
df_numerized=df
for col_name in df_numerized.columns:
    if(df_numerized[col_name].dtype=='object'):
        print(f"Converting column {col_name} to numeric")
        df_numerized[col_name]=df_numerized[col_name].astype('category')
        df_numerized[col_name]=df_numerized[col_name].cat.codes
        print(f"Column {col_name} converted to numeric")
df_numerized
```

```python
correlation_matrix=df_numerized.corr(method='pearson')
sns.heatmap(correlation_matrix, annot=True)
plt.title('Correlation matrice for  numeric features')
plt.xlabel('Movie features')
plt.ylabel('Movie features')
plt.show()
```
